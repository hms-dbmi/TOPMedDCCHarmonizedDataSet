```python
##READ IN ALL THE DATA FROM THE HARMONIZED FILES FROM DataSTAGE DCC AND GENERATE ONE MONOLITHIC FILE
def GenerateHarmonizedDF():
    hsubdir1 = "DCC_Harmonized/topmed_dcc_harmonized_atherosclerosis_v1/"
    hsubdir2 = "DCC_Harmonized/topmed_dcc_harmonized_baseline_common_covariates_v2/"
    hsubdir3 = "DCC_Harmonized/topmed_dcc_harmonized_blood_cell_count_v3/"
    hsubdir4 = "DCC_Harmonized/topmed_dcc_harmonized_blood_pressure_v1/"
    hsubdir5 = "DCC_Harmonized/topmed_dcc_harmonized_demographic_v3/"
    hsubdir6 = "DCC_Harmonized/topmed_dcc_harmonized_lipids_v3/"
    hsubdir7 = "DCC_Harmonized/topmed_dcc_harmonized_vte_v1/"
    
    fname1 = "topmed_dcc_harmonized_atherosclerosis_v1_eav.txt"
    fname2 = "topmed_dcc_harmonized_baseline_common_covariates_v2_eav.txt"
    fname3 = "topmed_dcc_harmonized_blood_cell_count_v3_eav.txt"
    fname4 = "topmed_dcc_harmonized_blood_pressure_v1_eav.txt"
    fname5 = "topmed_dcc_harmonized_demographic_v3_eav.txt"
    fname6 = "topmed_dcc_harmonized_lipids_v3_eav.txt"
    fname7 = "topmed_dcc_harmonized_vte_v1_eav.txt"
    
    hfile1 = baseDir + hsubdir1 + fname1
    hfile2 = baseDir + hsubdir2 + fname2
    hfile3 = baseDir + hsubdir3 + fname3
    hfile4 = baseDir + hsubdir4 + fname4
    hfile5 = baseDir + hsubdir5 + fname5
    hfile6 = baseDir + hsubdir6 + fname6
    hfile7 = baseDir + hsubdir7 + fname7
    
    AtherosclerosisDF = pd.read_csv(hfile1, comment="#", sep="\t")
    CommonCovariatesDF = pd.read_csv(hfile2, comment="#", sep="\t")
    BloodCellCountDF = pd.read_csv(hfile3, comment="#", sep="\t")
    BloodPressureDF = pd.read_csv(hfile4, comment="#", sep="\t")
    DemographicDF = pd.read_csv(hfile5, comment="#", sep="\t")
    LipidsDF = pd.read_csv(hfile6, comment="#", sep="\t")
    VteDF = pd.read_csv(hfile7, comment="#", sep="\t")
    
    HarmonizedDF = pd.concat([AtherosclerosisDF, CommonCovariatesDF])
    HarmonizedDF = pd.concat([HarmonizedDF, BloodCellCountDF])
    HarmonizedDF = pd.concat([HarmonizedDF, BloodPressureDF])
    HarmonizedDF = pd.concat([HarmonizedDF, DemographicDF])
    HarmonizedDF = pd.concat([HarmonizedDF, LipidsDF])
    HarmonizedDF = pd.concat([HarmonizedDF, VteDF])
    return HarmonizedDF
```

```python
##HELPER METHOD TO ASSEMBLE A DATA FRAME OF A STUDY SUBSET FROM THE HARMONIZED DATASETS
def GetStudySubsetFromHarmonizedData(HarmonizedDF, StudyName, Int64Flag=False):
    SubsetDF = HarmonizedDF.loc[HarmonizedDF['topmed_study'] == StudyName]
    SubsetDF = SubsetDF[[SubjectIdParamName, 'unique_subject_key', 'topmed_study']]
    SubsetDF = SubsetDF.drop_duplicates()
    print str(len(pd.unique(SubsetDF[SubjectIdParamName]))) + ' patients\' data has been harmonized from ' + StudyName + ' study'
    if Int64Flag:
        SubsetDF = SubsetDF.astype({'SUBJECT_ID': 'int64'}) ##Because SUBJECT_ID in CONSENT Tables is int64
    return SubsetDF    
```

```python
##HELPER METHOD TO GENERATE CONSENT DATA FRAME. DEFAULT VERSION 
def ProcessConsentFile(ConsentFileName, StudyName, ColumnToUseForSourceSubjectId=SSIdParamName):
    ConsentFile = baseDir + 'data_latest_version/' + ConsentFileName
    ConsentDF = pd.read_csv(ConsentFile, comment="#", sep="\t")
    
    if(ColumnToUseForSourceSubjectId != SSIdParamName):
        ConsentDF[SSIdParamName] = ConsentDF[ColumnToUseForSourceSubjectId]
    ConsentDF = ConsentDF[[DbGapParamName, SubjectIdParamName, ConsentParamName, SSIdParamName]]
    ConsentDF = ConsentDF.drop_duplicates()
    
    print str(len(pd.unique(ConsentDF[SubjectIdParamName]))) + ' ' + StudyName + ' patients in consent file'
    
    return ConsentDF
```

```python
##HELPER METHOD TO GENERATE CONSENT DATA FRAME WHERE SOME COLUMN RENAMING IS REQUIRED
def ProcessConsentFileWithColumnRenaming(ConsentFileName, StudyName, ColumnsToRename, ColumnToUseForSourceSubjectId=SSIdParamName):
    ConsentFile = baseDir + 'data_latest_version/' + ConsentFileName
    ConsentDF = pd.read_csv(ConsentFile, comment="#", sep="\t")
    
    if(ColumnToUseForSourceSubjectId != SSIdParamName):
        ConsentDF[SSIdParamName] = ConsentDF[ColumnToUseForSourceSubjectId]
    
    ConsentDF = ConsentDF.rename(columns = ColumnsToRename)    
    ConsentDF = ConsentDF[[DbGapParamName, SubjectIdParamName, ConsentParamName, SSIdParamName]]
    ConsentDF = ConsentDF.drop_duplicates()
    print str(len(pd.unique(ConsentDF[SubjectIdParamName]))) + ' ' + StudyName + ' patients in consent file'
    
    return ConsentDF
```

```python
##HELPER METHOD
def GenerateHarmonizedDataFrameWithConsentInfo(HarmonizedDF, ConsentDF, InputDF):
    ##INNER JOIN WITH SUBJECT TABLE(S) FROM dbGaP TO ADD CONSENT INFORMATION
    HarmonizedConsentDF = pd.merge(HarmonizedDF, ConsentDF, on='SUBJECT_ID', how='inner')

    print 'Found consent information for ' + str(len(pd.unique(HarmonizedConsentDF[SubjectIdParamName]))) + ' patients' 
    print 'Found Consent Levels: ' + str(pd.unique(HarmonizedConsentDF[ConsentParamName]))

    HarmonizedConsentDF = HarmonizedConsentDF[['unique_subject_key', 'SUBJECT_ID', 'dbGaP_Subject_ID', 'topmed_study', 'CONSENT', "SOURCE_SUBJECT_ID"]]
    print 
    
    OutputDF = pd.concat([InputDF, HarmonizedConsentDF])
    return OutputDF
```

```python
## COPDGene Study
def GetPatientsWithConsentInfoFromCopdGeneStudy(InputDF, HarmonizedDF):
    print 'COPDGene..'
    ##SELECT ROWS FROM HARMONIZED DF FOR COPD GENE STUDY
    CopdHarmonizedDF = GetStudySubsetFromHarmonizedData(HarmonizedDF, 'COPDGene')

    CopdConsentDF1 = ProcessConsentFile(CopdConsentFileName1, 'COPDGene')
    CopdConsentDF2 = ProcessConsentFile(CopdConsentFileName2, 'COPDGene', SubjectIdParamName)

    CopdConsentDF = pd.concat([CopdConsentDF1, CopdConsentDF2])

    OutputDF = GenerateHarmonizedDataFrameWithConsentInfo(CopdHarmonizedDF, CopdConsentDF, InputDF)
    return OutputDF
```

```python
## Framigham Heart Study
def GetPatientsWithConsentInfoFromFraminghamHeartStudy(InputDF, HarmonizedDF):
    print 'Framingham Heart Study..'
    ##SELECT ROWS FROM HARMONIZED DF FOR FRAMINGHAM STUDY
    FhsHarmonizedDF = GetStudySubsetFromHarmonizedData(HarmonizedDF, 'FHS', True)
    
    FhsConsentFile1 = baseDir + "data_latest_version/" + FhsConsentFileName1
    FhsConsentDF1 = pd.read_csv(FhsConsentFile1, comment="#", sep="\t")
    FhsConsentDF1 = FhsConsentDF1[[DbGapParamName, 'consent_1217', 'source_subjid']]
    FhsConsentDF1 = FhsConsentDF1.rename(columns={DbGapParamName:SubjectIdParamName, "consent_1217":ConsentParamName, "source_subjid":SSIdParamName})
    FhsConsentDF1[DbGapParamName] = FhsConsentDF1[SubjectIdParamName]
    FhsConsentDF1 = FhsConsentDF1.drop_duplicates()
    print str(len(pd.unique(FhsConsentDF1[SubjectIdParamName]))) + ' FHS patients in first file'

    FhsConsentFile2 = baseDir + 'data_latest_version/' + FhsConsentFileName2
    FhsConsentDF2 = pd.read_csv(FhsConsentFile2, comment="#", sep="\t")
    FhsConsentDF2 = FhsConsentDF2[[DbGapParamName, ConsentParamName]]
    FhsConsentDF2 = FhsConsentDF2.rename(columns={DbGapParamName:SubjectIdParamName})
    FhsConsentDF2[DbGapParamName] = FhsConsentDF2[SubjectIdParamName]
    FhsConsentDF2[SSIdParamName] = FhsConsentDF2[SubjectIdParamName]
    FhsConsentDF2 = FhsConsentDF2.drop_duplicates()

    print str(len(pd.unique(FhsConsentDF2[SubjectIdParamName]))) + ' FHS patients in Second TopMED file'

    FhsConsentDF = pd.concat([FhsConsentDF1, FhsConsentDF2])
    FhsConsentDF = FhsConsentDF.drop_duplicates()
    print str(len(pd.unique(FhsConsentDF[SubjectIdParamName]))) + ' FHS patients in Merged Consent file'
    
    OutputDF = GenerateHarmonizedDataFrameWithConsentInfo(FhsHarmonizedDF, FhsConsentDF, InputDF)
    return OutputDF
```

```python
##Genetic Epidemiology Network of Arteriopathy (GENOA)
def GetPatientsWithConsentInfoFromGenoaStudy(InputDF, HarmonizedDF):
    print "Genetic Epidemiology Network of Arteriopathy (GENOA)..."
    ##SELECT ROWS FROM HARMONIZED DF FOR Genetic Epidemiology Network of Arteriopathy (GENOA) STUDY
    GenoaHarmonizedDF = GetStudySubsetFromHarmonizedData(HarmonizedDF, 'GENOA')
    
    GenoaConsentDF = ProcessConsentFile(GenoaConsentFileName, 'GENOA', SubjectIdParamName)
    
    OutputDF = GenerateHarmonizedDataFrameWithConsentInfo(GenoaHarmonizedDF, GenoaConsentDF, InputDF)
    return OutputDF
```
